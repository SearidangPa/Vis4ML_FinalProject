### T-sne Projections of Feature Attributions by Different Methods

```python
attr_tsne_filename = '../Weights/Attr/attr_tsne.pkl'
attr_embedded_df = pd.read_pickle(attr_tsne_filename)
```

```python
brush = alt.selection_interval()

base = alt.Chart(attr_embedded_df).mark_circle().encode(
    opacity = alt.condition(brush, alt.value(1.0), alt.value(0.3))
).add_selection(
    brush
).properties(
    width=350,
    height=350
)

view_shap = base.encode(x='shap_x-tsne', y='shap_y-tsne', color = 'label:N',)
view_lime = base.encode(x='lime_x-tsne', y='lime_y-tsne', color = 'label:N',)
view_ig = base.encode(x='ig_x-tsne', y='ig_y-tsne', color = 'label:N',)
view_deepLift = base.encode(x='deepLift_x-tsne', y='deepLift_y-tsne', color = 'label:N',)

view1 = alt.hconcat(view_shap, view_lime)
view2 = alt.hconcat(view_ig, view_deepLift)
t_sne_attr_plot = alt.vconcat(view1, view2)
```
