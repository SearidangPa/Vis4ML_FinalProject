```python
import pandas as pd
import altair as alt
import warnings
warnings.filterwarnings('ignore')
```

```python
tsne_filename = '../Saved/Proj/tsne.pkl'
tsne_rank_filename = '../Saved/Proj/tsne_rank.pkl'
tsne_value_df = pd.read_pickle(tsne_filename).drop(['target'], axis=1)
tsne_rank_df = pd.read_pickle(tsne_rank_filename)
tsne_value_and_value = pd.concat([tsne_value_df, tsne_rank_df], axis = 1)
```

## Make the chart for T-sne of feature attribution values

```python
brush = alt.selection_interval()

base = alt.Chart(tsne_value_and_value).mark_circle().encode(
    opacity = alt.condition(brush, alt.value(1.0), alt.value(0.1))
).add_selection(
    brush
).properties(
    width=350,
    height=350
)

view_shap = base.encode(x='shap_x-tsne', y='shap_y-tsne', color = 'target:N',)
view_lime = base.encode(x='lime_x-tsne', y='lime_y-tsne', color = 'target:N',)
view_ig = base.encode(x='ig_x-tsne', y='ig_y-tsne', color = 'target:N',)
view_deepLift = base.encode(x='deepLift_x-tsne', y='deepLift_y-tsne', color = 'target:N',)
view_emb_data = base.encode(x='data_x-tsne', y='data_y-tsne', color = 'target:N',)

view1 = alt.hconcat(view_shap, view_lime)
view2 = alt.hconcat(view_ig, view_deepLift)
t_sne_attr_plot = alt.vconcat(view1, view2, title = 't-sne projection of feature attribute values')
```

## Make the chart for T-sne of signed feature rank 

```python
shap_rank_tsne = base.encode(x='shap_rank_x-tsne', y='shap_rank_y-tsne', color = 'target:N',)
lime_rank_tsne = base.encode(x='lime_rank_x-tsne', y='lime_rank_y-tsne', color = 'target:N',)
ig_rank_tsne = base.encode(x='ig_rank_x-tsne', y='ig_rank_y-tsne', color = 'target:N',)
deepLift_rank_tsne = base.encode(x='deepLift_rank_x-tsne', y='deepLift_rank_y-tsne', color = 'target:N',)
view1 = shap_rank_tsne | lime_rank_tsne
view2 = ig_rank_tsne | deepLift_rank_tsne 
t_sne_attr_rank_plot = alt.vconcat(view1, view2, title = 't-sne projection of signed feature values')
```

```python
t_sne_rank_and_value = alt.hconcat(t_sne_attr_plot, t_sne_attr_rank_plot)
```
